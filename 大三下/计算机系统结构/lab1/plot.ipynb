```python
x = []
y = []
with open('log_2.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        x.append(int(line[line.index('lock size = ') + 12:line.index('B, step =')]))
        y.append(float(line[line.index('end - begin = ') + 14:line.index('ms')]))

print(x, y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# plt.xscale('log')
plt.xlabel('log of array size(KB)')
plt.ylabel('time(ms)')
plt.xticks(np.log2(x), x)
plt.plot(np.log2(x), y)
```
